<div dir=rtl>
    از روش 1.5 iqr که سر کلاس بیان شد استفاده می‌کنم ولی
دقت کنید داده‌های reports و majorcards به دلیل محدودیت تعداد حالت
با این روش اطلاعات با ارزشی از دست می‌دهند بنابراین این روش را روی آن‌ها اجرا نمی‌کنم.

```python
fig, axes = plt.subplots(3,3, figsize=(20,20))
outlier_limit = {}
index = 0
for col in numerical:
    axes[index // 3][index % 3].set_title(col)
    sns.boxplot(df[col], ax=axes[index // 3][index % 3]);
    q1, q3 = df[col].quantile([0.25, 0.75])                 
    iqr = q3 - q1
    whisk1 = q1 - 1.5*iqr
    whisk2 = q3 + 1.5*iqr
    outlier_limit[col] = [whisk1, whisk2]
    index += 1
outlier_limit = pd.DataFrame(outlier_limit, index=['lower_bound', 'upper_bound'])
##omit for majorcards and repors
outlier_limit['majorcards'] = [0, df['majorcards'].max()]
outlier_limit['reports'] = [0, df['reports'].max()]
```

<div dir="rtl">
در برخی از ستون‌های اسمی عباراتی وجود دارند که معادل ناموجود بودن مقدار آن ردیف در ان ستون هستند. همچنین ستون‌های اسمی باینری نیز بهتر است به حالت True, False در بیایند.
        برای ستون‌های عددی نیز برخی مقادیر غیر مجاز یا ناموجود هستند(یعنی برای ستون‌هایی که باید مقدار مثبت داشته باشند، مقدار منفی ثبت شده است و یا داده‌ای بسیار بزرگ به عبارتی outlier وجود دارد) که باید به صورت None در بیایند.
    این ستون‌ها را تغییر دهید و دوباره نسبت مقادیر ناموجود هر ستون و همینطور اطلاعات آماری ستون های عددی تغییر یافته را محاسبه کنید.
    <br>
    (۵ نمره)
</div>

```python
upper_bounds = outlier_limit.loc['upper_bound']
df_copy = df.copy()

df_copy[numerical] = (df_copy[numerical] > upper_bounds) | (df[numerical] < 0)
df[numerical] = df[numerical].mask(df_copy[numerical], np.nan)
df.head()
```

```python
def derive_categorical(string):
    if string == 'yes' or string == '1' or string == 1:
        return True
    if string == 'no' or string == '0' or string == 0:
        return False
    if string == 'no-info':
        return np.nan
    return string

df['selfemp'] = df['selfemp'].map(derive_categorical, na_action='ignore')
df['owner'] = df['owner'].map(derive_categorical, na_action='ignore')
df.head()
```

```python
count = df.isna().sum()
percentage = count / len(df) * 100
print('number of nas in each column: ')
print(count)
print('percentage of nas in each column: ')
print(percentage)
```

```python
# Load and show some of data points
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
with np.load('mnist.zip') as data:
    arr, label = data['X'], data['y']
new_arr = arr.reshape(-1, 28 * 28)
mnist_df = pd.DataFrame(new_arr)
mnist_df['label'] = label
mnist_df.head()
```

```python
label = df['wealth']
new_df = df.drop(columns=['wealth'])
df_normalized = new_df - new_df.mean()
df_normalized.head()
```

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
cov_matrix = df_normalized.cov()
eigenvalues, eigenvectors, = np.linalg.eig(cov_matrix)
# notice that each column is an eigenvector
eig_pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]

eig_pairs.sort(reverse=True)
eigvalues_sort  = np.array([eig_pairs[index][0] for index in range(len(eigenvalues))])
eigvectors_sort = np.array([eig_pairs[index][1] for index in range(len(eigenvalues))])

print('Eigenvectors of Cov(X): ')
print(eigvectors_sort)
print('Eigenvalues of Cov(X):')
print(eigvalues_sort)
```

<div dir=rtl>
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
    <br>
    (۴ نمره)
</div>

```python
explained_variance = eigvalues_sort / eigvalues_sort.sum()
fig1, ax1 = plt.subplots()
fig1.set_size_inches(15, 10)
ax1.pie(explained_variance, labels = [i for i in range(1,14)], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

<div dir=rtl>
با حفظ حداقل ۵۰ درصد واریانس داده‌ها، ماتریس پرتو(projection matrix) را بدست آورید و تبدیل pca را بر روی دیتاست اجرا کنید.
    <br>
    (۵ نمره)
</div>

```python
reduced_data = np.array(eigvectors_sort[:2]).T
proj_data = df_normalized @ reduced_data
proj_data.columns = ['pc1', 'pc2']
proj_data.head()
```

<div dir=rtl>
با استفاده از مقادیر متناظر با pcهای اول و دوم هر داده در دیتاست حاصل شده، نموداری رسم کرده و تاثیر این روش را در جداسازی دو کلاس فقیر و ثروتمند تحلیل کنید.
    <br>
    (۴ نمره)
همانطور که بالاتر هم بیان شد می‌توان در اینجا هم دید که PCA
    منجر به تفکیک پذیری داده‌ها نشده است. پس موثر واقع نشده است.
</div>
